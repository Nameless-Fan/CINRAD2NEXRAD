2. Mutil files

```python
import os
from glob import glob
from radar.write.cinrad2nexrad import cinrad2nexrad

indir=os.path.abspath('D:/data/radar/NJUC/20220728/')
outdir='Z:/tmp_out/20220728/'
site_name_4s='NJUC'

os.makedirs(outdir,exist_ok=True)
inpaths = sorted(glob(os.path.join(indir,'*.bz2')))

for inpath in inpaths:
    fname=inpath.split('\\')[-1]
    outpath=os.path.join(outdir,fname+'.ar2v')
    try:
        cinrad2nexrad(inpath,outpath,site_name_4s)
    except:
        print(fname)
```
